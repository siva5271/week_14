```python
import pandas as pd
import numpy as np
from sklearn.model_selection import cross_val_score,train_test_split
from sklearn.tree import DecisionTreeRegressor
from sklearn.preprocessing import LabelEncoder
pd.set_option('display.max_columns',None)
```

```python
df=pd.read_csv('/home/siva/Downloads/py-master/ML/16_regularization/Melbourne_housing_FULL.csv')
df
```

```python
df.dropna(subset=['Price','Postcode'],inplace=True)
df.reset_index(inplace=True,drop=True)
df
```

```python
df.drop(['Car','Address','Date','Distance','CouncilArea','Lattitude'\
,'Longtitude','Regionname','Suburb','YearBuilt','BuildingArea'],\
axis=1,inplace=True)
```

```python
df['Bedroom2'].fillna(df['Bedroom2'].median(),inplace=True)
df['Bathroom'].fillna(df['Bathroom'].median(),inplace=True)
df['Landsize'].fillna(df['Landsize'].mean(),inplace=True)
df['Propertycount'].fillna(df['Propertycount'].mean(),inplace=True)
```

```python
df.info()
```

```python
X=df.drop(['Price'],axis=1)
y=df['Price']
```

```python
model=DecisionTreeRegressor()
score=cross_val_score(model,X,y,cv=100)
score.mean()
```
